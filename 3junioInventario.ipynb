Creamos un pequeño prpograma que descuente el inventario de una hamburgueseria


```python
inventario = 100 #definimos la variable inventario y le damos valor con un int de 100
print (f"El inventario contiene {inventario} hamburguesas") #mostramos al usuario el mensaje inicial
while inventario > 0: # Changed condition to allow loop to run
  if inventario <= 10:
    print ("El inventario esta casi vacio: CUIDADO") # Fixed typo
  num_hamburguesas = int(input("¿Cuantas hamburguesas quiere el cliente? "))
  if num_hamburguesas > inventario: # Fixed variable name from inventarios to inventario
    print("No hay suficientes hamburguesas. Solo hay",inventario,"hamburguesas disponibles")
  else:
    inventario -= num_hamburguesas
    print(f"El cliente ha pedido {num_hamburguesas} hamburguesas. El inventario ahora tiene {inventario}") # Fixed f-string syntax
    if inventario == 0:
      print("El inventario esta vacio.")
      break # Added break to exit loop when inventory is 0
```
